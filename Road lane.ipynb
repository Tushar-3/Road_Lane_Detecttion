```python
img=cv.imread("solidWhiteRight.jpg")
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray=cv.cvtColor(img,cv.COLOR_RGB2GRAY)
plt.imshow(img_gray,cmap="gray")
```

## Gaussian Blur

```python
img_blur=cv.GaussianBlur(img_gray,(5,5),0) ## by default 5,5 tthis one is the kernel size this one is  a hit and trail method
plt.imshow(img_blur)
```

```python
def get_vertice(image):
    rows,cols=image.shape[:2]
    bottom_left=[cols*0.1,rows*0.95]
    top_left=[cols*0.45,rows*0.6]
    bottom_right=[cols*0.9,rows*0.95]
    top_right=[cols*0.55,rows*0.6]
    ver=np.array([[bottom_left,top_left,bottom_right,top_right]],dtype=np.int32)
    return ver
```

```python
mask=np.zeros_like(img_canny)
plt.imshow(mask)
```

```python
com=cv.bitwise_and(img_canny,mask)
plt.imshow(com,cmap="gray")
```

```python
lines=cv.HoughLinesP(com,
             rho=6,
             theta=np.pi/60,
             threshold=140,
             lines=np.array([]),
             minLineLength=40,
             maxLineGap=5)
```
